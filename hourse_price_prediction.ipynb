```python
categorical_feature = new_df.select_dtypes(include='object')
```

### Convert categorical feature to numeric

```python
categorical_feature_copy = categorical_feature.copy()
for col in  categorical_feature_copy.columns :
    unique = categorical_feature_copy[col].unique()
    word_to_number = {word: number for number, word in enumerate(unique, 0)}
    for word, number in word_to_number.items():
        categorical_feature_copy[col] = categorical_feature_copy[col].replace(word, number)    
```

```python
categorical_feature = categorical_feature_copy
```

#### Normalizating the data to bring them in similar scale

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

# Identify numerical columns for normalization
numerical_feature = new_df.select_dtypes(include=['int64', 'float64'])
numerical_featureumns = numerical_feature.columns

# Use MinMaxScaler to normalize all numerical columns
scaler = MinMaxScaler()
numerical_feature[numerical_featureumns] = scaler.fit_transform(numerical_feature[numerical_featureumns])

# Display the DataFrame after normalization
print(numerical_feature)
```

```python
# Identify numerical columns for normalization
categorical_columns = categorical_feature.select_dtypes(include=['number']).columns

categorical_feature[categorical_columns] = scaler.fit_transform(categorical_feature[categorical_columns])

# Display the DataFrame after normalization
print(categorical_feature)
```

```python
new_df = pd.concat([numerical_feature,categorical_feature], axis=1)
```

## Feature Extraction
#### Using Mutual information for better feature selection and finding the dependencies between the features

```python
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif
from sklearn.model_selection import train_test_split

X = new_df.drop(columns=['SalePrice'], axis=1)
Y = new_df['SalePrice']

X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42 )

mutual_info_reg = mutual_info_regression(X_train.fillna(0),Y_train)
```

```python
mutual_info_reg
```

```python
# Create Feature Target Mutual Information Series
mi_series = pd.Series(mutual_info_reg)
mi_series.index = X_train.columns
mi_series.sort_values(ascending=False)
```

### Create feature target Mutual Information series 

```python
mi_series.sort_values(ascending=False).plot.bar(figsize=(20,8))
```

### Find out top 10 percentile features based on Mutual Information 

```python
from sklearn.feature_selection import SelectKBest, SelectPercentile
```

```python
sk_fs  = SelectKBest(score_func=mutual_info_regression, k=10)
sk_fs.fit(X_train.fillna(0), Y_train)
xs_xtrain =sk_fs.transform(X_train.fillna(0))
xs_xtest =sk_fs.transform(x_test.fillna(0))
```

```python
X_train.columns[sk_fs.get_support()]
```

```python
sk_classif_fs  = SelectPercentile(score_func=mutual_info_regression, percentile=20)
sk_classif_fs.fit_transform(X_train.fillna(0), Y_train)
xs_xtrain =sk_classif_fs.transform(X_train.fillna(0))
xs_xtest =sk_classif_fs.transform(x_test.fillna(0))
```

```python
X_train.columns[sk_classif_fs.get_support()]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(xs_xtrain, Y_train)
```

```python
y_pred = model.predict(xs_xtest)
```

```python
model.score(xs_xtest, y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
print("r2_score: ",r2_score(y_test, y_pred))
print("mean_squared_error:",mean_squared_error(y_test, y_pred))
print("mean_absolute_error:",mean_absolute_error(y_test, y_pred))
```
